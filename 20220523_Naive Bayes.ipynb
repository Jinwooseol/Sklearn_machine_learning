# Spam mail decision problem
$P(A|B)$ = $P(B|A)P(A) / (P(B|A)P(A) + P(B|A^{C})P(A^{C}))$
- '결제' 단어가 포함된 메일이 spam일 확률

$P(A)$: Spam mail일 확률  
$P(B)$: '결제'라는 단어가 포함될 확률  
$P(B|A)$: Spam mail 중 '결제' 단어가 포함된 확률  
$P(A^{C})$: Spam mail이 아닐 확률  
$P(B|A^{C})$: Spam mail이 아닌 메일 중 '결제'가 포함된 확률



```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X, y = load_iris(return_X_y = True, as_frame = True)
print(X)
print(y)
```

# Gaussian Naive Bayes

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)

gnb1 = GaussianNB()
gnb1.fit(X_train, y_train)
y_pred = gnb1.predict(X_test)
print(gnb1.score(X_test, y_test))   # Accuracy
print(y_test.values)
print(y_pred)
```

# Logistic Regression

```python
l1 = LogisticRegression(max_iter = 10000)
l1.fit(X_train, y_train)
print(l1.score(X_test, y_test))   # Accuracy
print(y_test.values)
print(l1.predict(X_test))
```

# Bernoulli Naive Bayes
- Not Good

```python
bnl1 = BernoulliNB()
bnl1.fit(X_train, y_train)
print(bnl1.score(X_test, y_test))   # Accuracy
print(y_test.values)
print(bnl1.predict(X_test))
```

# Confusion Matrix

image.png

Reference: https://manisha-sirsat.blogspot.com/2019/04/confusion-matrix.html

- Classifer의 성능을 확인하기 위한 Matrix
- **True Positive (TP)**: Actual positive and prediction is correct
- **False Positive (FP)**: Actual negative but prediction is wrong
- **False Negative (FN)**: Actual positive but prediction is wrong
- **True Negative (TN)**: Acture negative and prediction is correct

- **Sensitivity (Recall, True Positive Rate (TPR))**: $\frac{TP}{(TP + FN)}$
- **Specificity (False Positive Rate (FPR))**: $\frac{TN}{(TN + FP)}$
- **Accuracy**: $\frac{TP + TN}{(TP + TN + FP + FN)}$
- **Precision (Positive Predictive Value (PPV))**: $\frac{TP}{(TP + FP)}$
- **Type I Error**: FP
- **Type II Error**: FN
- **F1 score**: the harmonic mean of precision and sensitivity:  
  $F_{1} = 2 × \frac{precision × recall}{precision + recall}$  
  ${\displaystyle \mathrm {F} _{1}=2\times {\frac {\mathrm {PPV} \times \mathrm {TPR} }{\mathrm {PPV} +\mathrm {TPR} }}={\frac {2\mathrm {TP} }{2\mathrm {TP} +\mathrm {FP} +\mathrm {FN} }}}{\displaystyle \mathrm {F} _{1}=2\times {\frac {\mathrm {PPV} \times \mathrm {TPR} }{\mathrm {PPV} +\mathrm {TPR} }}={\frac {2\mathrm {TP} }{2\mathrm {TP} +\mathrm {FP} +\mathrm {FN} }}}$

# Calculate various types of score

```python
print('Accuracy score: ', metrics.accuracy_score(y_test, y_pred))
print('Counts of correct prediction: ', metrics.accuracy_score(y_test, y_pred, normalize = False))
print('Precision score: ', metrics.precision_score(y_test, y_pred, average = 'weighted'))
print('Recall score: ', metrics.recall_score(y_test, y_pred, average = None))
print('F1 score: ', metrics.f1_score(y_test, y_pred, average = None))
```

# ROC Curve
- 반드시 binary data만 가능
- y_score는 probability estimates or decision values가능
- Classifier의 성능을 비교하는데 사용

```python
y_pred_proba = gnb1.predict_proba(X_test)
#print(y_test[y_test != 2])
fpr, tpr, threshold = metrics.roc_curve(y_test[y_test != 0], y_pred_proba[:, 2][y_test != 0], pos_label = 2)
# pos_label -> positive label을 지정
print(fpr)
print(tpr)
print(threshold)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

_, axe = plt.subplots()
axe.plot(fpr, tpr)

sns.relplot(x = fpr, y = tpr)
```

# Area Under Curve (AUC-ROC)

```python
print(metrics.roc_auc_score(y_test[y_test != 0], y_pred_proba[:, 2][y_test != 0]))
```

# Application

# Breast Cancer class 분류하기
- Logistic Regression, Naive Bayes
- Features: mean radius, mean texture, mean area
- Target: Class(Malignant, Benign)

- Accuracy, precision, recall, f1 score 구하기
- Roc curve 구하기

```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
```

```python
print(load_breast_cancer().DESCR)
X, y = load_breast_cancer(return_X_y = True, as_frame = True)

X = X[['mean radius', 'mean texture', 'mean area']]
print(X.columns)
```

# Score function

```python
def total_score(y_actual, y_predict):
  print('Accuracy: ', accuracy_score(y_actual, y_predict))
  print('Precision: ', precision_score(y_actual, y_predict))
  print('Recall: ', recall_score(y_actual, y_predict))
  print('F1 score: ', f1_score(y_actual, y_predict))

def roc(y_actual, y_score, title):
  fpr, tpr, threshold = roc_curve(y_actual, y_score)
  _, axe = plt.subplots()
  axe.plot(fpr, tpr)
  axe.set_xlabel('False Positive Rate (FPR)')
  axe.set_ylabel('True Positive Rate (TPR)')
  auc_score = roc_auc_score(y_actual, y_score)
  axe.text(0.6, 0.8, 'AUC score: {:.3f}'.format(auc_score), fontsize = 15)
  axe.set_title(title, fontsize = 20)


def test_classify(fn, X_train, X_test, y_train, y_test):
  clf = fn().fit(X_train, y_train)
  y_pred = clf.predict(X_test)
```

# Logistic Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)

l1 = LogisticRegression().fit(X_train, y_train)
print('Score: ', l1.score(X_test, y_test))
l1_pred = l1.predict(X_test)

total_score(y_test, l1_pred)
roc(y_test, l1.predict_proba(X_test)[:, 1], 'Logistic Regression ROC Curve')
```

# Naive Bayes

```python
nb1 = GaussianNB().fit(X_train, y_train)

print('Score: ', nb1.score(X_test, y_test))
nb1_pred = nb1.predict(X_test)

total_score(y_test, nb1_pred)
roc(y_test, nb1.predict_proba(X_test)[:, 1], 'Naive Bayes ROC Curve')
```
